```python
import json

import requests
import pandas as pd
```

```python
url = 'https://data.colorado.gov/resource/4yw9-a5y6.json'
res = requests.get(url)
res.json()
```

```python
df = pd.json_normalize(res.json(), sep='__')
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.location__longitude
```

```python
df.variable.unique()
```

```python
cols = ['variable', 'amount', 'units', 'date_time', 'location__latitude', 'location__longitude']
df[df.variable == 'DISCHRG'][cols] \
.sort_values(by=['location__latitude', 'location__longitude'])
```

```python
cols = ['variable', 'amount', 'units', 'date_time', 'location__latitude', 'location__longitude']
df[df.variable == 'STORAGE'][cols] \
.sort_values(by=['location__latitude', 'location__longitude'])
```
